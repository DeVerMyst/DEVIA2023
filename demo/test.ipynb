```python
from matplotlib import pyplot as plt
nbpoint = 10000
N = np.random.randn(nbpoint)

plt.subplot(211)
plt.plot(N,'.')

interval = np.arange(-5,5)
hist = list()
for n in interval:
    plt.plot([0,10000],[n+.5,n+.5],':r')
    hist.append( np.sum( (N>(n-.5)) * (N<(n+.5)) ) )


x = interval
plt.xlim((0,nbpoint))
plt.subplot(212)
plt.bar(x,hist,1)
plt.grid()    
```

```python
from scipy.stats import norm
x2 = np.arange( np.min(x), np.max(x), .1)
no = norm.pdf(x2)
plt.bar(x,hist/np.sum(hist),1,label="histogramme")
plt.plot(x2,no ,'r',label="theorie")
plt.legend()
```

```python
plt.hist(N,bins=100, density=True,label="histogramme")
plt.plot(x2,no ,'r',label="theorie")
plt.legend()
```
