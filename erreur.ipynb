```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
a = .5
b = 0

x = np.arange(-10,10)
n = np.random.randn(len(x)) / 2
y = (a*x + b + n)**3

plt.plot(x,y,'or')
plt.grid()
```

Notre modele droite de type affine: 
Y = A X + B 

Les X ont les connais 

```python
x
```

On a une mesure de Y qu'on va appeler Y_m 

```python
erreur = list()
# for b in range(-10,10): 
a_list = np.linspace(-10,10,100)
for a in a_list: 
    model = a * x + b 
    e = np.mean( ( model - y)**2 ) 
    erreur.append(e)

plt.plot(erreur)        
plt.grid()

aest =  a_list[np.argmin(erreur)]
```

```python
plt.plot(x,y,'or')
plt.plot(x, aest*x , 'b',lw=2)
plt.grid()
```
